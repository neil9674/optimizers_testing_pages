# 2022 FRQ 4 — Grid Quest UI Game

This lesson turns **AP CSA 2022 FRQ 4** into a **Java UI game**.

## Main concept
**2D array traversal**
- Part A: fill every cell in a grid with valid random values
- Part B: scan each column and count how many are increasing



## FRQ Prompt

You are given this partial class:

```java
public class Data
{
    public static final int MAX = /* value not shown */;
    private int[][] grid;

    public void repopulate()
    { /* to be implemented in part (a) */ }

    public int countIncreasingCols()
    { /* to be implemented in part (b) */ }
}
```

### Part A — `repopulate`
Fill every element of `grid` with a random value such that:
- the value is between `1` and `MAX`, inclusive
- the value is divisible by `10`
- the value is **not** divisible by `100`

All valid values must have an equal chance of being generated.

### Part B — `countIncreasingCols`
Return the number of columns in `grid` that are in increasing order.

A column is increasing if every value after the first row is **greater than or equal to** the value above it.

window.GQState = window.GQState || {
  score: 0,
  health: 100
};

window.GQUpdate = function () {
  document.querySelectorAll(".gq-score-value").forEach(el => {
    el.textContent = String(window.GQState.score);
  });

  document.querySelectorAll(".gq-health-value").forEach(el => {
    el.textContent = window.GQState.health + "%";
  });

  document.querySelectorAll(".gq-health-fill").forEach(el => {
    el.style.width = window.GQState.health + "%";
  });

  let rank = "Array Rookie";
  if (window.GQState.score >= 5) rank = "Column Explorer";
  if (window.GQState.score >= 9) rank = "Grid Master";

  document.querySelectorAll(".gq-rank-value").forEach(el => {
    el.textContent = rank;
  });
};

window.GQAddScore = function(n){
  window.GQState.score += n;
  window.GQUpdate();
};

window.GQDamage = function(n){
  window.GQState.health = Math.max(0, window.GQState.health - n);
  window.GQUpdate();
};

window.GQHeal = function(n){
  window.GQState.health = Math.min(100, window.GQState.health + n);
  window.GQUpdate();
};

setTimeout(window.GQUpdate, 0);
</script>

          result.innerHTML = "<b>Signal miss.</b> That is not <code>grid[1][2]</code>.";

})();

    Random rand = new Random();

   for (int r = 0; r < grid.length; r++)

<script>
(function(){
  const root = document.getElementById("m2-shell");
  if (!root || root.dataset.ready) return;
  root.dataset.ready = "1";

  const values = [70,100,340,450,200,25,90,500];
  const valid = new Set([70,340,450,90]);
  const selected = new Set();
  let solved = false;

  const choices = root.querySelector("#m2-choices");
  const result = root.querySelector("#m2-result");

  values.forEach(v=>{
    const btn = document.createElement("button");
    btn.className = "gq-choice";
    btn.textContent = v;
    btn.onclick = () => {
      if (solved) return;
      if (selected.has(v)) {
        selected.delete(v);
        btn.classList.remove("selected");
      } else {
        selected.add(v);
        btn.classList.add("selected");
      }
    };
    choices.appendChild(btn);
  });

  root.querySelector("#m2-check").onclick = () => {
    if (solved) return;

    const ok = selected.size === valid.size && [...valid].every(v => selected.has(v));
    if (ok) {
      result.className = "gq-result good";
      result.innerHTML = "<b>Reactor stabilized.</b> Valid values are 70, 340, 450, and 90.";
      solved = true;
      window.GQAddScore(3);
      window.GQHeal(10);
    } else {
      result.className = "gq-result bad";
      result.innerHTML = "<b>Reactor overload warning.</b> Check divisibility by 10 and avoid multiples of 100.";
      window.GQDamage(12);
    }
  };

  root.querySelector("#m2-reset").onclick = () => {
    if (solved) return;
    selected.clear();
    choices.querySelectorAll(".gq-choice").forEach(btn => btn.classList.remove("selected"));
    result.className = "gq-result info";
    result.innerHTML = "Select all valid values, then press Check.";
  };

  window.GQUpdate();
})();
</script>

(function(){
  const root = document.getElementById("m4-shell");
  if (!root || root.dataset.ready) return;
  root.dataset.ready = "1";

  const rows = 3;
  const cols = 4;

  // Three phases. Player must infer the route from the feedback.
  // Phase 1 teaches row-major traversal.
  // Phase 2 teaches column traversal.
  // Phase 3 teaches increasing-column logic with a boolean flag.
  const phases = [
    {
      name: "Row Runner",
      runes: [
        {r:0,c:1,label:"R1"},
        {r:0,c:2,label:"R2"},
        {r:0,c:3,label:"R3"}
      ],
      start:{r:0,c:0},
      validateMove(prev, next, progress){
        // Must move right along row 0 only
        const expected = this.runes[progress];
        return next.r === expected.r && next.c === expected.c;
      },
      guide: "Row Runner: the dungeon favors the same order a nested loop would use across a row."
    },
    {
      name: "Column Scan",
      runes: [
        {r:1,c:0,label:"C1"},
        {r:2,c:0,label:"C2"},
        {r:0,c:1,label:"C3"}
      ],
      start:{r:0,c:0},
      validateMove(prev, next, progress){
        const expected = this.runes[progress];
        return next.r === expected.r && next.c === expected.c;
      },
      guide: "Column Scan: do not think in rows now. Scan vertically before moving to the next column."
    },
    {
      name: "Sentinel Check",
      runes: [
        {r:1,c:3,label:"S1"},
        {r:2,c:3,label:"S2"}
      ],
      start:{r:0,c:3},
      values:[20,25,30],
      validateMove(prev, next, progress){
        const expected = this.runes[progress];
        return next.r === expected.r && next.c === expected.c;
      },
      guide: "Sentinel Check: keep the boolean flag true only if values never decrease as you move down the column."
    }
  ];

  let phaseIndex = 0;
  let pos = {...phases[0].start};
  let progress = 0;
  let solvedAll = false;

  const grid = root.querySelector("#m4-grid");
  const result = root.querySelector("#m4-result");
  const phaseEl = root.querySelector("#m4-phase");
  const runesEl = root.querySelector("#m4-runes");
  const flagEl = root.querySelector("#m4-flag");
  const guideText = root.querySelector("#m4-guide-text");

  function currentPhase(){
    return phases[phaseIndex];
  }

  function isRune(r,c){
    return currentPhase().runes.find((x, idx) => idx >= progress && x.r===r && x.c===c);
  }

  function render(){
    grid.innerHTML = "";
    phaseEl.textContent = currentPhase().name;
    runesEl.textContent = `${progress} / ${currentPhase().runes.length}`;
    guideText.innerHTML = `<b>[] says:</b> ${currentPhase().guide}`;

    // flag display for final phase
    if (phaseIndex === 2) {
      flagEl.textContent = "increasing = true";
    } else {
      flagEl.textContent = "not active";
    }

    for(let r=0;r<rows;r++){
      for(let c=0;c<cols;c++){
        const cell = document.createElement("div");
        cell.className = "gq-cell";

        const rune = isRune(r,c);
        if (rune) {
          const runeTag = document.createElement("div");
          runeTag.className = "gq-rune";
          runeTag.textContent = "✦";
          cell.appendChild(runeTag);
        }

        const coord = document.createElement("div");
        coord.className = "coord";
        coord.textContent = `${r},${c}`;
        cell.appendChild(coord);

        if (r === pos.r && c === pos.c) {
          const player = document.createElement("div");
          player.className = "gq-player";
          player.textContent = "[]";
          cell.appendChild(player);
          cell.classList.add("visited");
        }

        // Show values subtly in final phase so it's more game-like, not target-obvious
        if (phaseIndex === 2 && c === 3) {
          const vals = [20,25,30];
          const v = document.createElement("div");
          v.style.position = "absolute";
          v.style.top = "8px";
          v.style.right = "8px";
          v.style.fontSize = "11px";
          v.style.color = "#cbd5e1";
          v.textContent = vals[r];
          cell.appendChild(v);
        }

        grid.appendChild(cell);
      }
    }
  }

  function setResult(type, html){
    result.className = "gq-result " + type;
    result.innerHTML = html;
  }

  function advancePhase(){
    phaseIndex++;
    if (phaseIndex >= phases.length) {
      solvedAll = true;
      setResult("good", "<b>Dungeon cleared.</b> You used movement to simulate traversal, flags, and checkpoint logic.");
      if (window.GQAddScore) window.GQAddScore(4);
      if (window.GQHeal) window.GQHeal(10);
      phaseIndex = phases.length - 1;
      return;
    }
    pos = {...currentPhase().start};
    progress = 0;
    setResult("good", `<b>Phase cleared.</b> Entering <b>${currentPhase().name}</b>.`);
    render();
  }

  function stepOn(next){
    const phase = currentPhase();

    // Rune check
    const ok = phase.validateMove(pos, next, progress);

    pos = next;
    render();

    if (ok) {
      progress++;
      setResult("good", "<b>Checkpoint found.</b> That movement matches the algorithm.");
      if (window.GQAddScore) window.GQAddScore(1);

      if (phaseIndex === 2) {
        // For the final phase, keep flag true because this column is increasing
        flagEl.textContent = "increasing = true";
      }

      if (progress >= phase.runes.length) {
        advancePhase();
      } else {
        runesEl.textContent = `${progress} / ${currentPhase().runes.length}`;
      }
    } else {
      // Wrong movement = algorithm error
      const idx = next.r * cols + next.c;
      const cell = grid.children[idx];
      if (cell) cell.classList.add("trap");

      if (phaseIndex === 0) {
        setResult("bad", "<b>Traversal error.</b> That is not row-major order.");
      } else if (phaseIndex === 1) {

        flagEl.textContent = "increasing = false?";
        setResult("bad", "<b>Sentinel warning.</b> That step does not match the check order for an increasing column.");
      }

      if (window.GQDamage) window.GQDamage(8);
    }
  }

    if (solvedAll) return;
    const nr = pos.r + dr;
    const nc = pos.c + dc;

    if (nr < 0 || nr >= rows || nc < 0 || nc >= cols) {
      setResult("bad", "<b>ArrayIndexOutOfBounds!</b> That move goes outside the array.");
      if (window.GQDamage) window.GQDamage(6);
      return;
    }

    stepOn({r:nr, c:nc});
  }

  function resetMission(){
    phaseIndex = 0;
    pos = {...phases[0].start};
    progress = 0;
    solvedAll = false;
    setResult("info", "Find the first rune. The board will punish incorrect traversal.");
    render();
  }

  root.querySelector("#m4-down").onclick = () => move(1,0);
  root.querySelector("#m4-left").onclick = () => move(0,-1);
  root.querySelector("#m4-right").onclick = () => move(0,1);
  root.querySelector("#m4-reset").onclick = resetMission;

  function keyHandler(e){
    if (!root.isConnected) {
      window.removeEventListener("keydown", keyHandler);
      return;
    }
    if (e.key === "ArrowUp") { e.preventDefault(); move(-1,0); }
    if (e.key === "ArrowDown") { e.preventDefault(); move(1,0); }
    if (e.key === "ArrowLeft") { e.preventDefault(); move(0,-1); }
    if (e.key === "ArrowRight") { e.preventDefault(); move(0,1); }
  }
  window.addEventListener("keydown", keyHandler);

  render();
  setResult("info", "Find the first rune. The board will punish incorrect traversal.");
  if (window.GQUpdate) window.GQUpdate();
})();

````java
// CODE_RUNNER: FRQ #4 (a) - repopulate()
import java.util.Arrays;
import java.util.Random;

public class Main {

    public static class Data {
        public static final int MAX = 500;
        private int[][] grid;

        public Data(int rows, int cols) {
            grid = new int[rows][cols];
        }

        /**
         * Fills all elements of grid with randomly generated values:
         * - between 1 and MAX inclusive
         * - divisible by 10
         * - NOT divisible by 100
         * All valid values have equal probability.
         */
        public void repopulate() {
            Random rand = new Random();

            int maxMultiple = MAX / 10;  // largest k such that 10*k <= MAX

            for (int r = 0; r < grid.length; r++) {
                for (int c = 0; c < grid[0].length; c++) {

                    int value;
                    do {
                        int k = rand.nextInt(maxMultiple) + 1; // 1..maxMultiple
                        value = 10 * k;
                    } while (value % 100 == 0);

                    grid[r][c] = value;
                }
            }
        }
    }


    public void driver() {
        Data d = new Data(3, 5);
        d.repopulate();

        System.out.println("Generated Grid:");
        for (int[] row : d.grid) {
            System.out.println(Arrays.toString(row));
        }
    }

    public static void main(String[] args) {
        Main m = new Main();
        m.driver();
    }
}

Main.main(null);
````

````java
// CODE_RUNNER: FRQ #4 (b) - countIncreasingCols()
import java.util.Arrays;

public class Main {

    public static class Data {
        private int[][] grid;

        public Data(int[][] g) {
            grid = g;
        }

        /**
         * Returns the number of columns in grid that are in increasing order.
         * A column is increasing if for every row r > 0:
         *   grid[r][c] >= grid[r-1][c]
         * A column with only one row is considered increasing.
         */
        public int countIncreasingCols() {
            int rows = grid.length;
            int cols = grid[0].length;
            int count = 0;

            for (int c = 0; c < cols; c++) {
                boolean increasing = true;

                for (int r = 1; r < rows; r++) {
                    if (grid[r][c] < grid[r - 1][c]) {
                        increasing = false;
                        break;
                    }
                }

                if (increasing) {
                    count++;
                }
            }

            return count;
        }
    }


    private static void printGrid(int[][] g) {
        for (int[] row : g) System.out.println(Arrays.toString(row));
    }

    public void driver() {
        int[][] example1 = {
            {10, 50, 40},
            {20, 40, 20},
            {30, 50, 30}
        };

        int[][] example2 = {
            {10, 540, 440, 440},
            {220, 450, 440, 190}
        };

        System.out.println("Example 1 Grid:");
        printGrid(example1);
        Data d1 = new Data(example1);
        System.out.println("countIncreasingCols() = " + d1.countIncreasingCols());
        System.out.println();

        System.out.println("Example 2 Grid:");
        printGrid(example2);
        Data d2 = new Data(example2);
        System.out.println("countIncreasingCols() = " + d2.countIncreasingCols());
        System.out.println();
    }

    public static void main(String[] args) {
        Main m = new Main();
        m.driver();
    }
}

Main.main(null);
````
