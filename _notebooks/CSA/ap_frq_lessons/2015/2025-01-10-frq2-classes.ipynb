## FRQ Question 2


Consider a guessing game in which a player tries to guess a hidden word. The hidden word contains only capital
letters and has a length known to the player. A guess contains only capital letters and has the same length as the
hidden word.

After a guess is made, the player is given a hint that is based on a comparison between the hidden word and the
guess. Each position in the hint contains a character that corresponds to the letter in the same position in the
guess. The following rules determine the characters that appear in the hint.

<table>
    <tr>
        <th>If the letter in the guess is … </th>
        <th>the corresponding character in the hint is</th>
    </tr>
    <tr>
        <td>also in the same position in the hidden word,</td>
        <td>the matching letter</td>
    </tr>
    <tr>
        <td>also in the hidden word, but in a different position</td>
        <td>"+"</td>
    </tr>
    <tr>
        <td>not in the hidden word, </td>
        <td>"*" </td>
    </tr>
<table>

The HiddenWord class will be used to represent the hidden word in the game. The hidden word is passed to
the constructor. The class contains a method, getHint, that takes a guess and produces a hint. 

For example, suppose the variable puzzle is declared as follows. 

HiddenWord puzzle = new HiddenWord("HARPS");

Write the complete HiddenWord class, including any necessary instance variables, its constructor, and the
method, getHint, described above. You may assume that the length of the guess is the same as the length of
the hidden word. 


```java
// CODE_RUNNER: HiddenWord
public class Main {

    // ===============================
    // 2015 FRQ #2 — HiddenWord class
    // ===============================
    public static class HiddenWord {

        // STUDENT CODE STARTS HERE
        private String word;
        // STUDENT CODE ENDS HERE

        /**
         * Constructs a HiddenWord object.
         * @param hWord the word to be guessed
         * Precondition: hWord contains only uppercase letters.
         */
        public HiddenWord(String hWord) {

            // STUDENT CODE STARTS HERE
            word = hWord;
            // STUDENT CODE ENDS HERE
        }

        /**
         * Returns a hint for guess.
         * @param guess the word guessed
         * Precondition: guess has the same length as the hidden word
         *               and contains only uppercase letters.
         */
        public String getHint(String guess) {

            // STUDENT CODE STARTS HERE

            String hint = "";

            for (int i = 0; i < guess.length(); i++) {
                String guessLetter = guess.substring(i, i + 1);

                if (guessLetter.equals(word.substring(i, i + 1))) {
                    hint += guessLetter;
                } else if (word.indexOf(guessLetter) != -1) {
                    hint += "+";
                } else {
                    hint += "*";
                }
            }

            return hint;

            // STUDENT CODE ENDS HERE
        }
    }

    // ===============================
    // Driver to prove it runs
    // ===============================
    public static void main(String[] args) {

        HiddenWord puzzle = new HiddenWord("HARPS");

        System.out.println(puzzle.getHint("AAAAA")); // +****
        System.out.println(puzzle.getHint("HELLO")); // H****
        System.out.println(puzzle.getHint("HEART")); // H*++*
        System.out.println(puzzle.getHint("HARMS")); // HAR*S
    }
}

// Required by runner
Main.main(null);
```

<h2>Interactive HiddenWord Game</h2>
<p>Play this mini-game to experience the <code>getHint()</code> method in action! The game implements the same logic you need to code for the FRQ.</p>
<p><strong>How it works:</strong></p>
<ul>
    <li>You have 6 attempts to guess the 5-letter word</li>
    <li>Green letters are in the correct position (same as returning the letter)</li>
    <li>Yellow letters are in the word but wrong position (same as returning <code>+</code>)</li>
    <li>Gray letters are not in the word (same as returning <code>*</code>)</li>
</ul>

```java
%%js

// GAME_RUNNER: Play the HiddenWord guessing game! Experience the getHint() logic from the FRQ. | hide_edit: true

import GameControl from "/assets/js/GameEnginev1.1/essentials/GameControl.js";
import GameObject from "/assets/js/GameEnginev1.1/essentials/GameObject.js";

class HiddenWordGameObject extends GameObject {
    constructor(data, gameEnv) {
        super(gameEnv);
        this.words = [
            'HARPS', 'GRAPE', 'CLASH', 'BRAIN', 'PLANT',
            'STORM', 'LIGHT', 'SPACE', 'DREAM', 'WORLD',
            'CRANE', 'SLATE', 'TRACE', 'CRATE', 'STARE',
            'SIREN', 'PARSE', 'ATONE', 'AUDIO', 'SOUTH',
            'BEACH', 'BLOCK', 'CHAIR', 'DANCE', 'EARTH',
            'FLAME', 'GHOST', 'HOUSE', 'JUICE', 'KNIFE',
            'LEMON', 'MOUNT', 'NIGHT', 'OCEAN', 'PEARL',
            'QUEEN', 'ROVER', 'STONE', 'THINK', 'ULTRA',
            'VALOR', 'WHEAT', 'YOUTH', 'AMBER', 'BLAZE',
            'CHARM', 'DRIFT', 'EAGLE', 'FORGE', 'GLORY',
            'HASTE', 'INPUT', 'JUMBO', 'KNELT', 'LUNAR',
            'MAPLE', 'NERVE', 'ORBIT', 'PIXEL', 'QUEST',
            'REIGN', 'SWIFT', 'TIMER', 'UNITY', 'VIVID'
        ];
        this.targetWord = '';
        this.currentRow = 0;
        this.currentCol = 0;
        this.maxRows = 6;
        this.maxCols = 5;
        this.grid = [];
        this.currentGuess = '';
        this.gameOver = false;
        this.message = '';

        this.canvas = document.createElement("canvas");
        this.canvas.width = gameEnv.innerWidth;
        this.canvas.height = gameEnv.innerHeight;
        this.ctx = this.canvas.getContext("2d");
        this.gameEnv.container.appendChild(this.canvas);

        this.resetGame();

        this.handleKeyDownBound = this.handleKeyDown.bind(this);
        document.addEventListener('keydown', this.handleKeyDownBound);
    }

    resetGame() {
        this.targetWord = this.words[Math.floor(Math.random() * this.words.length)];
        this.currentRow = 0;
        this.currentCol = 0;
        this.currentGuess = '';
        this.gameOver = false;
        this.message = '';
        this.grid = [];
        for (let r = 0; r < this.maxRows; r++) {
            this.grid.push([]);
            for (let c = 0; c < this.maxCols; c++) {
                this.grid[r].push({ letter: '', status: '' });
            }
        }
    }

    /** Same logic as the FRQ getHint() method */
    getHint(guess) {
        let result = [];
        for (let i = 0; i < guess.length; i++) {
            const guessLetter = guess.charAt(i);
            if (guessLetter === this.targetWord.charAt(i)) {
                result.push({ letter: guessLetter, status: 'correct' });
            } else if (this.targetWord.indexOf(guessLetter) !== -1) {
                result.push({ letter: guessLetter, status: 'present' });
            } else {
                result.push({ letter: guessLetter, status: 'absent' });
            }
        }
        return result;
    }

    handleKeyDown(e) {
        if (this.gameOver) {
            if (e.key === 'Enter') {
                this.resetGame();
            }
            return;
        }

        if (e.key === 'Backspace') {
            if (this.currentCol > 0) {
                this.currentCol--;
                this.currentGuess = this.currentGuess.slice(0, -1);
                this.grid[this.currentRow][this.currentCol] = { letter: '', status: '' };
            }
        } else if (e.key === 'Enter') {
            if (this.currentGuess.length === this.maxCols) {
                const hint = this.getHint(this.currentGuess);
                for (let i = 0; i < this.maxCols; i++) {
                    this.grid[this.currentRow][i] = hint[i];
                }
                if (this.currentGuess === this.targetWord) {
                    this.message = 'You won! getHint() returned the full word! Press Enter for new game.';
                    this.gameOver = true;
                } else {
                    this.currentRow++;
                    this.currentCol = 0;
                    this.currentGuess = '';
                    if (this.currentRow >= this.maxRows) {
                        this.message = 'Game Over! The word was: ' + this.targetWord + '. Press Enter for new game.';
                        this.gameOver = true;
                    }
                }
            } else {
                this.message = 'Enter a 5-letter word!';
            }
        } else if (/^[a-zA-Z]$/.test(e.key) && this.currentCol < this.maxCols) {
            const letter = e.key.toUpperCase();
            this.grid[this.currentRow][this.currentCol] = { letter: letter, status: '' };
            this.currentGuess += letter;
            this.currentCol++;
            this.message = '';
        }
    }

    update() {
        this.draw();
    }

    draw() {
        const ctx = this.ctx;
        const w = this.canvas.width;
        const h = this.canvas.height;
        ctx.clearRect(0, 0, w, h);

        // Background
        ctx.fillStyle = '#f9f9f9';
        ctx.fillRect(0, 0, w, h);

        const cellSize = 52;
        const gap = 5;
        const gridW = this.maxCols * (cellSize + gap) - gap;
        const gridH = this.maxRows * (cellSize + gap) - gap;
        const startX = (w - gridW) / 2;
        const startY = 45;

        // Title
        ctx.fillStyle = '#333';
        ctx.font = 'bold 20px Arial';
        ctx.textAlign = 'center';
        ctx.textBaseline = 'top';
        ctx.fillText('HiddenWord Game', w / 2, 10);

        // Draw grid
        for (let r = 0; r < this.maxRows; r++) {
            for (let c = 0; c < this.maxCols; c++) {
                const cell = this.grid[r][c];
                const x = startX + c * (cellSize + gap);
                const y = startY + r * (cellSize + gap);

                if (cell.status === 'correct') {
                    ctx.fillStyle = '#6aaa64';
                } else if (cell.status === 'present') {
                    ctx.fillStyle = '#c9b458';
                } else if (cell.status === 'absent') {
                    ctx.fillStyle = '#787c7e';
                } else {
                    ctx.fillStyle = '#ffffff';
                }
                ctx.fillRect(x, y, cellSize, cellSize);

                ctx.strokeStyle = (r === this.currentRow && !this.gameOver) ? '#878a8c' : '#d3d6da';
                ctx.lineWidth = 2;
                ctx.strokeRect(x, y, cellSize, cellSize);

                if (cell.letter) {
                    ctx.fillStyle = cell.status ? '#ffffff' : '#333333';
                    ctx.font = 'bold 28px Arial';
                    ctx.textAlign = 'center';
                    ctx.textBaseline = 'middle';
                    ctx.fillText(cell.letter, x + cellSize / 2, y + cellSize / 2);
                }
            }
        }

        // Message
        if (this.message) {
            ctx.fillStyle = '#333';
            ctx.font = '14px Arial';
            ctx.textAlign = 'center';
            ctx.textBaseline = 'top';
            ctx.fillText(this.message, w / 2, startY + gridH + 12);
        }

        // Legend
        const legendY = startY + gridH + 40;
        ctx.font = 'bold 13px Arial';
        ctx.textAlign = 'left';
        ctx.fillStyle = '#333';
        ctx.fillText('Legend (Same as FRQ getHint!):', startX, legendY);

        const items = [
            { color: '#6aaa64', text: 'Correct position (returns letter)' },
            { color: '#c9b458', text: 'In word, wrong position (returns +)' },
            { color: '#787c7e', text: 'Not in word (returns *)' }
        ];
        ctx.font = '12px Arial';
        items.forEach((item, i) => {
            const iy = legendY + 18 + i * 20;
            ctx.fillStyle = item.color;
            ctx.fillRect(startX, iy, 14, 14);
            ctx.fillStyle = '#333';
            ctx.fillText(item.text, startX + 20, iy + 11);
        });
    }

    resize() {
        // Fixed size — no resize needed
    }

    destroy() {
        document.removeEventListener('keydown', this.handleKeyDownBound);
        if (this.canvas && this.canvas.parentNode) {
            this.canvas.parentNode.removeChild(this.canvas);
        }
    }
}

class GameLevelHiddenWord {
    constructor(gameEnv) {
        this.classes = [
            { class: HiddenWordGameObject, data: {} }
        ];
    }
}

export const gameLevelClasses = [GameLevelHiddenWord];
export { GameControl };
```
