import java.util.ArrayList;
import java.util.Arrays;
public class Main
{
    private ArrayList<String> wordList;
    public Main(ArrayList<String> words)
    {
        wordList = words;
    }
    public boolean isWordChain()
    {
        for (int i = 1; i < wordList.size(); i++)
        {
            if (!wordList.get(i).contains(wordList.get(i - 1)))
            {
                return false;
            }
        }
        return true;
    }

import java.util.ArrayList;
import java.util.Arrays;
public class Main
{
    private ArrayList<String> wordList;
    public Main(ArrayList<String> words)
    {
        wordList = words;
    }
    /**
     * Returns an ArrayList<String> based on strings from wordList that start
     * with target. Each element of the returned ArrayList has had
     * the initial occurrence of target removed.
     * Postconditions: wordList is unchanged.
     *                 Items appear in the returned list in the same order as they appear in wordList.
     */
    public ArrayList<String> createList(String target)
    {
        ArrayList<String> result = new ArrayList<>();
        for (String w : wordList)
        {
            if (w.startsWith(target))
            {
                result.add(w.substring(target.length()));
            }
        }
        return result;
    }
    // Test driver
    public static void main(String[] args)
    {
        ArrayList<String> words =
            new ArrayList<>(Arrays.asList("catch", "bobcat", "catchacat", "cat", "at"));
        Main checker = new Main(words);
        System.out.println("Original wordList: " + words);
        System.out.println();
        ArrayList<String> result1 = checker.createList("cat");
        System.out.println("createList(\"cat\")");
        System.out.println("Expected: [ch, chacat, ]");
        System.out.println("Result: " + result1);
        System.out.println();
        ArrayList<String> result2 = checker.createList("catch");
        System.out.println("createList(\"catch\")");
        System.out.println("Expected: [, acat]");
        System.out.println("Result: " + result2);
        System.out.println();
        ArrayList<String> result3 = checker.createList("dog");
        System.out.println("createList(\"dog\")");
        System.out.println("Expected: []");
        System.out.println("Result: " + result3);
    }

        guessedDisplay.textContent = gameState.guessed.length > 0 ? gameState.guessed.join(', ') : '--';

        document.getElementById('hangman-wrong').textContent = gameState.wrongCount;

        const hintEl = document.getElementById('hangman-hint');
        hintEl.textContent = '💡 ' + (HINTS[gameState.word] || 'Guess the word!');
    }

    function guessLetter(letter) {
        if (gameState.gameOver || gameState.guessed.includes(letter)) return;

        gameState.guessed.push(letter);

        if (!gameState.word.includes(letter)) {
            gameState.wrongCount++;
        }

        checkGameEnd();
        updateDisplay();
        drawHangman(gameState.wrongCount);
        updateKeyboard();
    }

    function checkGameEnd() {
        const won = gameState.word.split('').every(letter => gameState.guessed.includes(letter));
        const lost = gameState.wrongCount >= 6;

        if (won) {
            gameState.gameOver = true;
            gameState.won = true;
            document.getElementById('hangman-message').textContent = '🎉 You Won! The word was ' + gameState.word;
            document.getElementById('hangman-message').className = 'hangman-message--win';
        } else if (lost) {
            gameState.gameOver = true;
            gameState.won = false;
            document.getElementById('hangman-message').textContent = '💀 Game Over! The word was ' + gameState.word;
            document.getElementById('hangman-message').className = 'hangman-message--lose';
        } else {
            document.getElementById('hangman-message').textContent = '';
        }
    }

    function updateKeyboard() {
        const keyboard = document.getElementById('hangman-keyboard');
        const buttons = keyboard.querySelectorAll('button');
        buttons.forEach(btn => {
            const letter = btn.textContent;
            if (gameState.guessed.includes(letter)) {
                btn.disabled = true;
            }
        });
    }

    window.hangmanNewGame = function() {
        const wordData = selectRandomWord();
        gameState = {
            word: wordData.word,
            guessed: [],
            wrongCount: 0,
            gameOver: false,
            won: false
        };

        // Rebuild keyboard
        const keyboard = document.getElementById('hangman-keyboard');
        keyboard.innerHTML = '';
        for (let i = 0; i < 26; i++) {
            const letter = String.fromCharCode(65 + i);
            const btn = document.createElement('button');
            btn.className = 'hangman-btn-letter';
            btn.textContent = letter;
            btn.onclick = () => {
                guessLetter(letter);
            };
            keyboard.appendChild(btn);
        }

        document.getElementById('hangman-message').textContent = '';
        updateDisplay();
        drawHangman(0);
    };

    // Initialize on load
    if (document.readyState === 'loading') {
        document.addEventListener('DOMContentLoaded', hangmanNewGame);
    } else {
        hangmanNewGame();
    }
})();
</script>
