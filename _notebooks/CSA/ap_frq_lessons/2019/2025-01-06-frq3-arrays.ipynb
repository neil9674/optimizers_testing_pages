````python
// CODE_RUNNER: Part (a) -- Complete the getDelimitersList method

import java.util.ArrayList;

public class Delimiters {
    private String openDel;
    private String closeDel;

    public Delimiters(String open, String close) {
        openDel = open;
        closeDel = close;
    }

    /** Part (a): Returns an ArrayList of delimiters from the array tokens */
    public ArrayList<String> getDelimitersList(String[] tokens) {
        ArrayList<String> result = new ArrayList<String>();
        for (int i = 0; i < tokens.length; i++) {
            if (tokens[i].equals(openDel) || tokens[i].equals(closeDel)) {
                result.add(tokens[i]);
            }
        }
        return result;
    }

    public static void main(String[] args) {
        // Test with parentheses
        Delimiters d1 = new Delimiters("(", ")");
        String[] tokens1 = {"(", "x + y", ")", "* 5"};
        System.out.println("Test 1: " + d1.getDelimitersList(tokens1));
        // Expected: [(, )]

        // Test with HTML tags
        Delimiters d2 = new Delimiters("<q>", "</q>");
        String[] tokens2 = {"<q>", "yy", "</q>", "zz", "</q>"};
        System.out.println("Test 2: " + d2.getDelimitersList(tokens2));
        // Expected: [<q>, </q>, </q>]
    }
}

Delimiters.main(null);
````

````python
// CODE_RUNNER: Part (B) -- Complete the isBalanced method

import java.util.ArrayList;

public class Delimiters {
    private String openDel;
    private String closeDel;

    public Delimiters(String open, String close) {
        openDel = open;
        closeDel = close;
    }

    /** Part (a): Extract delimiters from tokens */
    public ArrayList<String> getDelimitersList(String[] tokens) {
        ArrayList<String> result = new ArrayList<String>();
        for (int i = 0; i < tokens.length; i++) {
            if (tokens[i].equals(openDel) || tokens[i].equals(closeDel)) {
                result.add(tokens[i]);
            }
        }
        return result;
    }

    /** Part (b): Check if delimiters are balanced */
    public boolean isBalanced(ArrayList<String> delimiters) {
        int openCount = 0;
        int closeCount = 0;
        for (int i = 0; i < delimiters.size(); i++) {
            if (delimiters.get(i).equals(openDel)) {
                openCount++;
            } else {
                closeCount++;
            }
            // Rule 1: At no point should closes exceed opens
            if (closeCount > openCount) {
                return false;
            }
        }
        // Rule 2: Final count of opens must equal closes
        return openCount == closeCount;
    }

    public static void main(String[] args) {
        Delimiters d = new Delimiters("<sup>", "</sup>");

        // Test 1: Balanced -- expected true
        ArrayList<String> t1 = new ArrayList<>();
        t1.add("<sup>"); t1.add("<sup>"); t1.add("</sup>");
        t1.add("<sup>"); t1.add("</sup>"); t1.add("</sup>");
        System.out.println("Test 1 (expect true):  " + d.isBalanced(t1));

        // Test 2: Close before open -- expected false (Rule 1)
        ArrayList<String> t2 = new ArrayList<>();
        t2.add("<sup>"); t2.add("</sup>"); t2.add("</sup>"); t2.add("<sup>");
        System.out.println("Test 2 (expect false): " + d.isBalanced(t2));

        // Test 3: Single close -- expected false (Rule 1)
        ArrayList<String> t3 = new ArrayList<>();
        t3.add("</sup>");
        System.out.println("Test 3 (expect false): " + d.isBalanced(t3));

        // Test 4: Unequal counts -- expected false (Rule 2)
        ArrayList<String> t4 = new ArrayList<>();
        t4.add("<sup>"); t4.add("<sup>"); t4.add("</sup>");
        System.out.println("Test 4 (expect false): " + d.isBalanced(t4));
    }
}

Delimiters.main(null);
````

## Delimiter Balance Runner

Use the **arrow keys** (or **A/D**) to move your collector left and right. Delimiters fall from the sky. You must catch them **in the correct order** to build a balanced sequence. Catching a delimiter that breaks the balance costs a life. The sequence must be balanced when the wave ends -- if not, you lose a life and retry. Survive all waves to win.

This game reinforces the exact logic of `isBalanced`: you must track open vs close counts in real time, and order matters.

<script>
/**
 * Delimiter Balance Runner
 *
 * Architecture mirrors GameEnv / GameControl / GameObject:
 *   GameEnv  -- canvas, dimensions, theme color refresh
 *   Player   -- keyboard-controlled collector (GameObject)
 *   Token    -- falling delimiter (GameObject)
 *   GameControl -- spawner, collision detection, game loop
 *
 * Educational tie-in: the player physically experiences the isBalanced
 * algorithm.  openCount is shown live.  Catching a close when
 * openCount == 0 is an immediate "Rule 1 violation."
 */
(function() {
    const canvas = document.getElementById('frq-game-canvas');
    if (!canvas) return;
    const ctx = canvas.getContext('2d');

    // ---- GameEnv ----
    const GameEnv = {
        canvas: canvas,
        ctx: ctx,
        W: canvas.width,
        H: canvas.height,
        colors: { bg:'#0b1220', text:'#e6eef8', accent:'#3b82f6',
                  green:'#4ADE80', red:'#E53E3E', orange:'#ED8936',
                  teal:'#38B2AC', muted:'#8899aa', panel:'#1a2233' },
        refresh: function() {
            var s = getComputedStyle(document.documentElement);
            this.colors.bg     = s.getPropertyValue('--pref-bg-color').trim()     || this.colors.bg;
            this.colors.text   = s.getPropertyValue('--pref-text-color').trim()   || this.colors.text;
            this.colors.accent = s.getPropertyValue('--pref-accent-color').trim() || this.colors.accent;
        }
    };

    // ---- Input ----
    var keys = {};
    document.addEventListener('keydown', function(e) {
        keys[e.key] = true;
        if (['ArrowLeft','ArrowRight','a','d',' '].indexOf(e.key) !== -1) e.preventDefault();
    });
    document.addEventListener('keyup', function(e) { keys[e.key] = false; });

    // ---- Player (GameObject) ----
    function Player() {
        this.w = 64;
        this.h = 24;
        this.x = GameEnv.W / 2 - this.w / 2;
        this.y = GameEnv.H - 50;
        this.speed = 5;
        this.bobTimer = 0;
        this.catchFlash = 0;      // > 0 means flash green
        this.catchFlashBad = 0;   // > 0 means flash red
    }
    Player.prototype.update = function() {
        if (keys['ArrowLeft'] || keys['a']) this.x -= this.speed;
        if (keys['ArrowRight'] || keys['d']) this.x += this.speed;
        if (this.x < 0) this.x = 0;
        if (this.x + this.w > GameEnv.W) this.x = GameEnv.W - this.w;
        this.bobTimer += 0.06;
        if (this.catchFlash > 0) this.catchFlash--;
        if (this.catchFlashBad > 0) this.catchFlashBad--;
    };
    Player.prototype.draw = function() {
        var c = GameEnv.colors;
        var bob = Math.sin(this.bobTimer) * 2;
        var yy = this.y + bob;

        // Glow
        if (this.catchFlash > 0) {
            ctx.shadowColor = c.green;
            ctx.shadowBlur = 18;
        } else if (this.catchFlashBad > 0) {
            ctx.shadowColor = c.red;
            ctx.shadowBlur = 18;
        }

        // Platform body
        ctx.fillStyle = this.catchFlash > 0 ? c.green : (this.catchFlashBad > 0 ? c.red : c.accent);
        ctx.beginPath();
        ctx.roundRect(this.x, yy, this.w, this.h, 8);
        ctx.fill();

        // Inner highlight
        ctx.fillStyle = 'rgba(255,255,255,0.15)';
        ctx.beginPath();
        ctx.roundRect(this.x + 4, yy + 3, this.w - 8, 8, 4);
        ctx.fill();

        ctx.shadowColor = 'transparent';
        ctx.shadowBlur = 0;

        // Label
        ctx.fillStyle = c.bg;
        ctx.font = 'bold 11px monospace';
        ctx.textAlign = 'center';
        ctx.textBaseline = 'middle';
        ctx.fillText('CATCH', this.x + this.w / 2, yy + this.h / 2);
        ctx.textAlign = 'start';
        ctx.textBaseline = 'alphabetic';
    };

    // ---- Token (falling GameObject) ----
    function Token(text, type, x, vy) {
        this.text = text;
        this.type = type; // 'open' | 'close' | 'trash'
        this.x = x;
        this.y = -30;
        this.vy = vy;
        this.w = 0;
        this.h = 30;
        this.alive = true;
        this.caught = false;
        this.rotation = (Math.random() - 0.5) * 0.02;
        this.angle = 0;
    }
    Token.prototype.update = function() {
        if (!this.alive) return;
        this.y += this.vy;
        this.angle += this.rotation;
        if (this.y > GameEnv.H + 40) this.alive = false; // off screen
    };
    Token.prototype.draw = function() {
        if (!this.alive) return;
        var c = GameEnv.colors;
        ctx.save();
        ctx.font = 'bold 16px monospace';
        this.w = ctx.measureText(this.text).width + 24;
        var cx = this.x + this.w / 2;
        var cy = this.y + this.h / 2;
        ctx.translate(cx, cy);
        ctx.rotate(this.angle);
        ctx.translate(-cx, -cy);

        // Shadow
        ctx.fillStyle = 'rgba(0,0,0,0.3)';
        ctx.beginPath();
        ctx.roundRect(this.x + 2, this.y + 2, this.w, this.h, 6);
        ctx.fill();

        // Body
        var bgColor, borderColor;
        if (this.type === 'open') {
            bgColor = c.accent + '44';
            borderColor = c.accent;
        } else if (this.type === 'close') {
            bgColor = c.orange + '44';
            borderColor = c.orange;
        } else {
            bgColor = c.red + '33';
            borderColor = c.red;
        }
        ctx.fillStyle = bgColor;
        ctx.strokeStyle = borderColor;
        ctx.lineWidth = 2;
        ctx.beginPath();
        ctx.roundRect(this.x, this.y, this.w, this.h, 6);
        ctx.fill();
        ctx.stroke();

        // Text
        ctx.fillStyle = c.text;
        ctx.font = 'bold 15px monospace';
        ctx.textAlign = 'center';
        ctx.textBaseline = 'middle';
        ctx.fillText(this.text, this.x + this.w / 2, this.y + this.h / 2);
        ctx.textAlign = 'start';
        ctx.textBaseline = 'alphabetic';
        ctx.restore();
    };

    // ---- Particles ----
    var particles = [];
    function spawnParticles(x, y, color, count) {
        for (var i = 0; i < count; i++) {
            particles.push({
                x: x, y: y,
                vx: (Math.random() - 0.5) * 4,
                vy: -Math.random() * 3 - 1,
                life: 30 + Math.random() * 20,
                maxLife: 50,
                color: color,
                size: 2 + Math.random() * 3
            });
        }
    }
    function updateParticles() {
        for (var i = particles.length - 1; i >= 0; i--) {
            var p = particles[i];
            p.x += p.vx;
            p.y += p.vy;
            p.vy += 0.08;
            p.life--;
            if (p.life <= 0) particles.splice(i, 1);
        }
    }
    function drawParticles() {
        particles.forEach(function(p) {
            var alpha = p.life / p.maxLife;
            ctx.globalAlpha = alpha;
            ctx.fillStyle = p.color;
            ctx.beginPath();
            ctx.arc(p.x, p.y, p.size, 0, Math.PI * 2);
            ctx.fill();
        });
        ctx.globalAlpha = 1;
    }

    // ---- Floating text ----
    var floatingTexts = [];
    function addFloatingText(x, y, text, color) {
        floatingTexts.push({ x: x, y: y, text: text, color: color, life: 50 });
    }
    function updateFloatingTexts() {
        for (var i = floatingTexts.length - 1; i >= 0; i--) {
            floatingTexts[i].y -= 1.2;
            floatingTexts[i].life--;
            if (floatingTexts[i].life <= 0) floatingTexts.splice(i, 1);
        }
    }
    function drawFloatingTexts() {
        floatingTexts.forEach(function(ft) {
            ctx.globalAlpha = ft.life / 50;
            ctx.fillStyle = ft.color;
            ctx.font = 'bold 14px monospace';
            ctx.textAlign = 'center';
            ctx.fillText(ft.text, ft.x, ft.y);
            ctx.textAlign = 'start';
        });
        ctx.globalAlpha = 1;
    }

    // ---- Game State ----
    var player = new Player();
    var tokenList = [];
    var score = 0, lives = 3, wave = 1;
    var openCount = 0;
    var sequence = [];
    var gameState = 'playing'; // 'playing' | 'waveComplete' | 'waveFailed' | 'gameOver' | 'victory'
    var stateTimer = 0;
    var spawnTimer = 0;
    var spawnInterval = 70; // frames between spawns
    var tokensToSpawn = 0;
    var tokensSpawnedThisWave = 0;
    var waveTokenCount = 0;
    var starField = [];

    // Star background
    for (var si = 0; si < 80; si++) {
        starField.push({
            x: Math.random() * GameEnv.W,
            y: Math.random() * GameEnv.H,
            size: 0.5 + Math.random() * 1.5,
            speed: 0.1 + Math.random() * 0.3,
            twinkle: Math.random() * Math.PI * 2
        });
    }

    var DELIM_PAIRS = [
        ['(', ')'], ['[', ']'], ['{', '}'],
        ['<B>', '</B>'], ['<i>', '</i>']
    ];
    var TRASH_TOKENS = ['#', '@', '&', '~', '%', '^', '!'];

    function buildWave(waveNum) {
        // Each wave defines a sequence of tokens to drop.
        // Contains a solvable balanced set plus distractors.
        var plan = [];
        var pair = DELIM_PAIRS[(waveNum - 1) % DELIM_PAIRS.length];
        var numPairs = Math.min(waveNum + 1, 5);
        // Build a balanced target: interleave opens then closes with some mixing
        var opens = [];
        var closes = [];
        for (var pi = 0; pi < numPairs; pi++) {
            opens.push(pair[0]);
            closes.push(pair[1]);
        }
        // Interleave into a solvable order
        var balanced = [];
        var oi = 0, ci = 0;
        while (oi < opens.length || ci < closes.length) {
            // Bias toward open first, sprinkle closes when we have opens banked
            if (oi < opens.length && (ci >= closes.length || Math.random() < 0.6 || oi <= ci)) {
                balanced.push({ text: opens[oi], type: 'open' }); oi++;
            } else {
                balanced.push({ text: closes[ci], type: 'close' }); ci++;
            }
        }
        plan = plan.concat(balanced);

        // Add distractors -- wrong-type delimiters and trash
        var distractorCount = Math.min(waveNum, 4);
        for (var di = 0; di < distractorCount; di++) {
            if (Math.random() < 0.5) {
                var wrongPair = DELIM_PAIRS[(waveNum + di) % DELIM_PAIRS.length];
                plan.push({ text: wrongPair[Math.random() < 0.5 ? 0 : 1], type: 'trash' });
            } else {
                plan.push({ text: TRASH_TOKENS[Math.floor(Math.random() * TRASH_TOKENS.length)], type: 'trash' });
            }
        }

        // Shuffle
        for (var shi = plan.length - 1; shi > 0; shi--) {
            var j = Math.floor(Math.random() * (shi + 1));
            var tmp = plan[shi]; plan[shi] = plan[j]; plan[j] = tmp;
        }
        return plan;
    }

    var wavePlan = [];

    function startWave() {
        tokenList = [];
        openCount = 0;
        sequence = [];
        spawnTimer = 0;
        tokensSpawnedThisWave = 0;
        wavePlan = buildWave(wave);
        waveTokenCount = wavePlan.length;
        tokensToSpawn = wavePlan.length;
        spawnInterval = Math.max(35, 70 - wave * 5);
        gameState = 'playing';
        updateHUD();
    }

    function updateHUD() {
        document.getElementById('frq-g-score').textContent = score;
        document.getElementById('frq-g-lives').textContent = lives;
        document.getElementById('frq-g-level').textContent = wave;
        document.getElementById('frq-g-counter').textContent = openCount;
        var seqStr = sequence.length > 0 ? sequence.join(' ') : '--';
        if (seqStr.length > 30) seqStr = '...' + seqStr.slice(-27);
        document.getElementById('frq-g-seq').textContent = seqStr;
    }

    // ---- Collision ----
    function checkCollisions() {
        var c = GameEnv.colors;
        for (var i = 0; i < tokenList.length; i++) {
            var t = tokenList[i];
            if (!t.alive || t.caught) continue;
            // AABB overlap with player
            if (t.y + t.h >= player.y && t.y <= player.y + player.h &&
                t.x + t.w >= player.x && t.x <= player.x + player.w) {
                t.alive = false;
                t.caught = true;

                if (t.type === 'trash') {
                    // Trash -- lose a life
                    lives--;
                    player.catchFlashBad = 15;
                    spawnParticles(t.x + t.w / 2, t.y, c.red, 12);
                    addFloatingText(t.x + t.w / 2, t.y - 10, 'WRONG', c.red);
                } else if (t.type === 'close') {
                    // Check Rule 1: would this make openCount negative?
                    if (openCount <= 0) {
                        lives--;
                        player.catchFlashBad = 15;
                        spawnParticles(t.x + t.w / 2, t.y, c.red, 12);
                        addFloatingText(t.x + t.w / 2, t.y - 10, 'RULE 1', c.red);
                    } else {
                        openCount--;
                        sequence.push(t.text);
                        score += 10 * wave;
                        player.catchFlash = 12;
                        spawnParticles(t.x + t.w / 2, t.y, c.green, 8);
                        addFloatingText(t.x + t.w / 2, t.y - 10, '+' + (10 * wave), c.green);
                    }
                } else {
                    // Open
                    openCount++;
                    sequence.push(t.text);
                    score += 5 * wave;
                    player.catchFlash = 12;
                    spawnParticles(t.x + t.w / 2, t.y, c.accent, 8);
                    addFloatingText(t.x + t.w / 2, t.y - 10, '+' + (5 * wave), c.accent);
                }
                updateHUD();
                if (lives <= 0) { gameState = 'gameOver'; stateTimer = 120; }
                break;
            }
        }
    }

    // ---- Check wave end ----
    function checkWaveEnd() {
        // Wave ends when all tokens have been spawned and none are alive
        if (tokensSpawnedThisWave >= waveTokenCount) {
            var anyAlive = false;
            for (var i = 0; i < tokenList.length; i++) {
                if (tokenList[i].alive) { anyAlive = true; break; }
            }
            if (!anyAlive) {
                // Check if balanced (Rule 2)
                if (openCount === 0 && sequence.length > 0) {
                    stateTimer = 90;
                    gameState = 'waveComplete';
                    score += 50 * wave; // bonus
                    updateHUD();
                } else if (sequence.length === 0) {
                    // Caught nothing valid -- fail
                    lives--;
                    stateTimer = 90;
                    gameState = 'waveFailed';
                    updateHUD();
                    if (lives <= 0) { gameState = 'gameOver'; stateTimer = 120; }
                } else {
                    // Unbalanced -- Rule 2 failure
                    lives--;
                    stateTimer = 90;
                    gameState = 'waveFailed';
                    updateHUD();
                    if (lives <= 0) { gameState = 'gameOver'; stateTimer = 120; }
                }
            }
        }
    }

    // ---- Draw helpers ----
    function drawStars() {
        var c = GameEnv.colors;
        starField.forEach(function(st) {
            st.y += st.speed;
            st.twinkle += 0.03;
            if (st.y > GameEnv.H) { st.y = 0; st.x = Math.random() * GameEnv.W; }
            ctx.globalAlpha = 0.3 + Math.sin(st.twinkle) * 0.3;
            ctx.fillStyle = c.text;
            ctx.beginPath();
            ctx.arc(st.x, st.y, st.size, 0, Math.PI * 2);
            ctx.fill();
        });
        ctx.globalAlpha = 1;
    }

    function drawOpenCountBar() {
        var c = GameEnv.colors;
        var barX = GameEnv.W - 30;
        var barH = GameEnv.H - 100;
        var barY = 50;
        var barW = 16;

        // Background track
        ctx.fillStyle = c.panel;
        ctx.beginPath();
        ctx.roundRect(barX, barY, barW, barH, 4);
        ctx.fill();

        // Fill proportional to openCount (max ~6)
        var maxDisplay = 6;
        var fillRatio = Math.min(openCount / maxDisplay, 1);
        var fillH = fillRatio * barH;

        var fillColor = openCount === 0 ? c.green : (openCount > 3 ? c.orange : c.accent);
        ctx.fillStyle = fillColor;
        ctx.beginPath();
        ctx.roundRect(barX, barY + barH - fillH, barW, fillH, 4);
        ctx.fill();

        // Label
        ctx.fillStyle = c.text;
        ctx.font = 'bold 12px monospace';
        ctx.textAlign = 'center';
        ctx.fillText(openCount.toString(), barX + barW / 2, barY - 8);
        ctx.font = '9px monospace';
        ctx.fillStyle = c.muted;
        ctx.fillText('open', barX + barW / 2, barY - 20);
        ctx.textAlign = 'start';
    }

    function drawOverlay(title, subtitle, color) {
        ctx.fillStyle = 'rgba(0,0,0,0.7)';
        ctx.fillRect(0, 0, GameEnv.W, GameEnv.H);
        ctx.fillStyle = color;
        ctx.font = 'bold 26px sans-serif';
        ctx.textAlign = 'center';
        ctx.fillText(title, GameEnv.W / 2, GameEnv.H / 2 - 12);
        ctx.fillStyle = GameEnv.colors.muted;
        ctx.font = '15px sans-serif';
        ctx.fillText(subtitle, GameEnv.W / 2, GameEnv.H / 2 + 18);
        ctx.textAlign = 'start';
    }

    // ---- GameControl: Main Loop ----
    function gameLoop() {
        GameEnv.refresh();
        var c = GameEnv.colors;

        // Background
        ctx.fillStyle = c.bg;
        ctx.fillRect(0, 0, GameEnv.W, GameEnv.H);
        drawStars();

        // Danger zone
        ctx.fillStyle = c.red + '12';
        ctx.fillRect(0, GameEnv.H - 55, GameEnv.W, 55);

        if (gameState === 'playing') {
            // Spawn tokens
            spawnTimer++;
            if (spawnTimer >= spawnInterval && tokensSpawnedThisWave < waveTokenCount) {
                var def = wavePlan[tokensSpawnedThisWave];
                var tx = 20 + Math.random() * (GameEnv.W - 80);
                var tv = 1.0 + wave * 0.25 + Math.random() * 0.5;
                tokenList.push(new Token(def.text, def.type, tx, tv));
                tokensSpawnedThisWave++;
                spawnTimer = 0;
            }

            // Update
            player.update();
            for (var i = 0; i < tokenList.length; i++) tokenList[i].update();
            // Remove dead off-screen tokens
            for (var j = tokenList.length - 1; j >= 0; j--) {
                if (!tokenList[j].alive && !tokenList[j].caught) tokenList.splice(j, 1);
            }
            updateParticles();
            updateFloatingTexts();
            checkCollisions();
            checkWaveEnd();

            // Draw
            for (var k = 0; k < tokenList.length; k++) tokenList[k].draw();
            player.draw();
            drawParticles();
            drawFloatingTexts();
            drawOpenCountBar();

        } else if (gameState === 'waveComplete') {
            player.draw();
            drawOpenCountBar();
            drawOverlay('Wave ' + wave + ' Complete', 'Balanced sequence achieved -- Score: ' + score, c.green);
            stateTimer--;
            if (stateTimer <= 0) {
                wave++;
                if (wave > 8) {
                    gameState = 'victory';
                    stateTimer = 200;
                } else {
                    startWave();
                }
            }

        } else if (gameState === 'waveFailed') {
            player.draw();
            var reason = openCount > 0 ? 'Rule 2: openCount != 0' : 'Not enough delimiters caught';
            drawOverlay('Wave Failed', reason + ' -- Lives: ' + lives, c.orange);
            stateTimer--;
            if (stateTimer <= 0) startWave();

        } else if (gameState === 'gameOver') {
            drawOverlay('Game Over', 'Final Score: ' + score + '  --  Click or press Space to retry', c.red);
            stateTimer--;

        } else if (gameState === 'victory') {
            drawOverlay('All Waves Cleared', 'Final Score: ' + score + '  --  Click or press Space to play again', c.green);
        }

        requestAnimationFrame(gameLoop);
    }

    // ---- Restart handling ----
    function restart() {
        score = 0; lives = 3; wave = 1;
        particles = []; floatingTexts = [];
        player = new Player();
        startWave();
    }

    canvas.addEventListener('click', function() {
        if (gameState === 'gameOver' || gameState === 'victory') restart();
    });
    document.addEventListener('keydown', function(e) {
        if (e.key === ' ' && (gameState === 'gameOver' || gameState === 'victory')) {
            e.preventDefault();
            restart();
        }
    });

    // ---- Start ----
    startWave();
    gameLoop();
})();
</script>

        const container = document.getElementById('frq-quiz-container');
        container.innerHTML = '';
        QUIZ.forEach((q, qi) => {
            let html = '<div class="frq-quiz__question" id="frq-q-' + qi + '">';
            html += '<div class="q-number">Question ' + (qi+1) + ' of ' + QUIZ.length + '</div>';
            html += '<div class="q-text">' + q.q + '</div>';
            html += '<div class="frq-quiz__choices">';
            q.choices.forEach((c, ci) => {
                html += '<div class="frq-quiz__choice" onclick="frqQuizAnswer(' + qi + ',' + ci + ')" id="frq-q' + qi + '-c' + ci + '">' + c + '</div>';
            });
            html += '</div>';
            html += '<div class="frq-quiz__explanation" id="frq-q' + qi + '-explain"></div>';
            html += '</div>';
            container.innerHTML += html;
        });
    }

    window.frqQuizAnswer = function(qi, ci) {
        const q = QUIZ[qi];
        const choices = document.querySelectorAll('#frq-q-' + qi + ' .frq-quiz__choice');
        const explain = document.getElementById('frq-q' + qi + '-explain');

        if (choices[0].classList.contains('frq-quiz__choice--correct') ||
            choices[0].classList.contains('frq-quiz__choice--incorrect') ||
            choices[0].classList.contains('frq-quiz__choice--disabled')) return;

        choices.forEach(c => c.classList.add('frq-quiz__choice--disabled'));

        if (ci === q.correct) {
            choices[ci].classList.add('frq-quiz__choice--correct');
            explain.className = 'frq-quiz__explanation frq-quiz__explanation--correct';
            explain.innerHTML = '<strong>Correct.</strong> ' + q.explain;
            quizScore++;
        } else {
            choices[ci].classList.add('frq-quiz__choice--incorrect');
            choices[q.correct].classList.add('frq-quiz__choice--correct');
            explain.className = 'frq-quiz__explanation frq-quiz__explanation--incorrect';
            explain.innerHTML = '<strong>Incorrect.</strong> ' + q.explain;
        }

        answered++;
        if (answered >= QUIZ.length) {
            const scoreDiv = document.getElementById('frq-quiz-score');
            scoreDiv.classList.remove('frq-quiz__score--hidden');
            document.getElementById('frq-quiz-score-val').textContent = quizScore + '/' + QUIZ.length;
            document.getElementById('frq-quiz-bar').style.width = (quizScore / QUIZ.length * 100) + '%';
        }
    };

    if (document.readyState === 'loading') document.addEventListener('DOMContentLoaded', renderQuiz);
    else renderQuiz();
})();
